**THEORY QUESTIONS**

1. What is a random variable in probability theory?
- A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment. It can be either discrete (taking countable values) or continuous (taking infinite values within a range).

2. What are the types of random variables?
- There are two main types:

  - Discrete Random Variable: Takes distinct, separate values (e.g., the number of heads in a coin toss).

  - Continuous Random Variable: Takes any value within a given range (e.g., height, temperature).

3. What is the difference between discrete and continuous distributions?
- Discrete Distribution: Defined for countable values (e.g., binomial, Poisson).

- Continuous Distribution: Defined for uncountable values over an interval (e.g., normal, exponential).

4. What are probability distribution functions (PDF)?
- The Probability Distribution Function (PDF) describes how probabilities are distributed over different values of a random variable. For discrete variables, it is called the Probability Mass Function (PMF).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- PDF (or PMF for discrete data) gives the probability of a specific value.

- CDF gives the probability that a random variable is less than or equal to a given value.

6. What is a discrete uniform distribution?
- A discrete uniform distribution is where all possible outcomes have equal probability. Example: Rolling a fair die (each number 1–6 has equal probability).

7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution models a binary outcome (success or failure).

- It has a probability parameter p, where:

    - P(Success) = p

    - P(Failure) = 1 - p

- Example: A single coin flip with p = 0.5.

11. What are the characteristics of a normal distribution?
- Bell-shaped and symmetric.

- Mean, median, and mode are equal.

- Defined by mean (μ) and standard deviation (σ).

- 68-95-99.7 rule: 68% within 1σ, 95% within 2σ, 99.7% within 3σ.

12. What is the standard normal distribution, and why is it important?
- A standard normal distribution is a normal distribution with:

  - Mean (μ) = 0

  - Standard deviation (σ) = 1

- It is used in Z-scores and hypothesis testing.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population distribution.

- Importance:

  - Enables normal approximation for various problems.

  - Justifies hypothesis testing and confidence intervals.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the discrete uniform distribution
x = np.arange(1, 7)  # Values (like rolling a fair die)
pmf = stats.randint.pmf(x, 1, 7)

# Plot the PMF
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of a Discrete Uniform Distribution")
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import scipy.stats as stats

def bernoulli_pdf(p, x):
    return stats.bernoulli.pmf(x, p)

# Example usage
p = 0.5  # Probability of success
x = [0, 1]  # Possible outcomes
pdf_values = bernoulli_pdf(p, x)

print("Bernoulli PDF values:", pdf_values)
```

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a binomial distribution
n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(data, bins=np.arange(12)-0.5, density=True, alpha=0.7, color='green', edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

5. Create a Poisson distribution and visualize it using Python.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define Poisson distribution with mean (lambda)
lambda_ = 5
x = np.arange(0, 15)
pmf = stats.poisson.pmf(x, lambda_)

# Plot PMF
plt.bar(x, pmf, color='purple', alpha=0.7)
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Poisson Distribution (λ=5)")
plt.show()
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Define the distribution
x = np.arange(1, 7)
cdf = stats.randint.cdf(x, 1, 7)

# Plot CDF
plt.plot(x, cdf, marker="o", linestyle="-", color="red")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of a Discrete Uniform Distribution")
plt.grid()
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import scipy.stats as stats

def calculate_z_scores(data):
    return stats.zscore(data)

# Generate random dataset
data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color="cyan", edgecolor="black")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal (exponential) distribution
population = np.random.exponential(scale=2, size=10000)

# Take multiple sample means
sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color="brown", edgecolor="black")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
from statsmodels.stats.weightstats import ztest

data = np.random.normal(50, 10, 30)
z_stat, p_value = ztest(data, value=50)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import scipy.stats as stats

sample = np.random.normal(50, 10, 30)
mean = np.mean(sample)
sem = stats.sem(sample)
ci = stats.t.interval(0.95, len(sample)-1, loc=mean, scale=sem)

print("95% Confidence Interval:", ci)
```
